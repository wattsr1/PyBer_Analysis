```python
# Import dependancies
import matplotlib.pyplot as plt
```

```python
# Set the x axis to a list  of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US Dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Set the x limit between Jan and June.
#plt.xlim(0, 5)
# Add a grid
plt.grid()
# create a title.
plt.title("PyBer Fare by Month")
#Add the legend.
plt.legend()
# Change figure size
plt.figure(figsize=(20, 10))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,alpha=0.8, edgecolor="Black", linewidth=2, c="skyblue", s = [i *5 for i in y_axis], label='Boston')
ax.invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")

```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, colors=colors, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, colors=colors, startangle=90, explode=explode_values, counterclock=False, autopct='%.1f%%', labels=x_axis)

plt.show()
```
